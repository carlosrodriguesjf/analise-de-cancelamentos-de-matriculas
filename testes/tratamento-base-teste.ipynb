```python
# importações

import pandas as pd
from datetime import datetime
import warnings

warnings.filterwarnings('ignore')
```

```python
# importado a base de dados
df = pd.read_excel(r'C:\projetos\Analise-de-cancelamentos-de-matricula-de-universidade\dados\base-de-dados.xlsx')
```

```python
# filtrando somente os valores protocolados
df_filtrado = df.loc[df['Status']=='PROTOCOLADO',:]
df_filtrado.head(5)
```

```python
# Função para reordenar as palavras
def reorder_tipo_ingresso(text):
    palavras_chave = ['SISU', 'PISM', 'Vestibular']
    for palavra in palavras_chave:
        if palavra in text and not text.startswith(palavra):
            # Move a palavra chave para o início
            partes = text.split(' - ', 1)  # Divide a string no primeiro ' - ' (se houver)
            return f"{palavra} - {partes[1]}" if len(partes) > 1 else palavra
    return text  # Retorna o texto original se já começar com a palavra chave

# Aplica a função para reordenar
df_filtrado['tipo_ingresso'] = df_filtrado['tipo_ingresso'].apply(reorder_tipo_ingresso)
```

```python
# preenchendo os espaços vazios da coluna Motivo para cancelamento
df_filtrado['Motivo para cancelamento'] = df_filtrado['Motivo para cancelamento'].fillna('Não especificado')
```

```python
# separando a coluna curso
df_filtrado[['codigo', 'curso']] = df_filtrado['curso'].str.split(' - ', n = 1, expand=True)
```

```python
# separando os dados da coluna tipo_ingresso
df_filtrado[['tipo_ingresso', 'observacao']] = df_filtrado['tipo_ingresso'].str.rsplit(' - ', n = 1, expand=True)
```

```python
def dividir_tipo_ingresso(text):
    # Remove qualquer traço presente no texto
    text = text.replace('-', '')
    
    # Substitui "Vest", "VEST", ou "Vestibular" por "VESTIBULAR"
    text = text.replace('Vest', 'VESTIBULAR').replace('VEST', 'VESTIBULAR').replace('Vestibular', 'VESTIBULAR')
    
    # Lista das palavras-chave a serem buscadas
    palavras_chave = ['SISU', 'PISM', 'VESTIBULAR', 'Vestibular EAD', 'Vest EAD']
    
    # Itera sobre as palavras-chave
    for palavra in palavras_chave:
        # Verifica se a palavra-chave está no texto
        if palavra in text:
            # Encontra o índice da palavra-chave e divide a string
            pos = text.find(palavra)
            # Divide a string na posição logo após a palavra-chave
            return text[:pos+len(palavra)], text[pos+len(palavra):].strip().upper()
    
    # Caso não encontre nenhuma palavra-chave, retorna o texto original
    return text, ''

# Aplica a função para dividir a coluna
df_filtrado[['tipo_ingresso', 'grupo']] = df_filtrado['tipo_ingresso'].apply(lambda x: pd.Series(dividir_tipo_ingresso(x)))
```

```python
# Remove as palavras 'IBULARIBULAR' e ' - ' da coluna grupo
df_filtrado['grupo'] = df_filtrado['grupo'].apply(lambda x: x.replace('IBULARIBULAR', '').replace(' - ', ''))
```

```python
# Remove as palavras 'IBULAR' e ' - ' da coluna grupo
df_filtrado['grupo'] = df_filtrado['grupo'].apply(lambda x: x.replace('IBULAR ', '').replace(' - ', ''))
```

```python
# Remove as palavras 'ENEM'  da coluna grupo
df_filtrado['grupo'] = df_filtrado['grupo'].apply(lambda x: x.replace(' ENEM', ''))
```

```python
# Remove o espaço no início das palavras
df_filtrado['grupo'] = df_filtrado['grupo'].apply(lambda x: x.lstrip())
```

```python
#Remove as palavras 'IBULAR' e ' - ' da coluna descricao
df_filtrado['grupo'] = df_filtrado['grupo'].apply(lambda x: x.replace('EAD', ''))
```

```python
df_filtrado['grupo'] = df_filtrado['grupo'].replace('', 'Não se aplica')
```

```python
# removendo colunas que não serão utilizadas na análise
df_filtrado.drop(['Status','Matricula', 'nome', 'classificacao','codigo','observacao','habilitacao','semestre_ingresso'], axis=1, inplace=True)
```

```python
#alterando o tipo da variavel ano_ingresso
df_filtrado['ano_ingresso'] = df_filtrado['ano_ingresso'].astype(int).astype('str')
```

```python
# alterando e uniformizando o nome das colunas
df_filtrado.rename(columns={'Motivo para cancelamento': 'motivo_cancelamento', 'Campus': 'campus', 'Data protocolo':'data_protocolo'}, inplace=True)
```

```python
# alterando o separador de decimais da variavel pontuação
#Substituir vírgulas por pontos apenas se houver vírgula
df_filtrado['pontuacao'] = df_filtrado['pontuacao'].apply(
    lambda x: str(x).replace(',', '.') if isinstance(x, str) or ',' in str(x) else str(x)
)

# Caso queira converter a coluna para float após a substituição
df_filtrado['pontuacao'] = df_filtrado['pontuacao'].astype(float)
```

```python
# Converter todas as entradas para o formato datetime
df_filtrado['data_protocolo'] = pd.to_datetime(df_filtrado['data_protocolo'], errors='coerce')

# Converter para o formato desejado: DD/MM/YYYY
df_filtrado['data_protocolo'] = df_filtrado['data_protocolo'].dt.strftime('%d/%m/%Y')
```

```python
# Converter todas as entradas para o formato datetime
df_filtrado['data_nascimento'] = pd.to_datetime(df_filtrado['data_nascimento'], errors='coerce')

# Converter para o formato desejado: DD/MM/YYYY
df_filtrado['data_nascimento'] = df_filtrado['data_nascimento'].dt.strftime('%d/%m/%Y')
```
